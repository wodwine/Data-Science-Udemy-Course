```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observations = 1000
x1 = np.random.uniform(low=-10, high=10, size=(observations,1)) # x1
x2 = np.random.uniform(-10, 10, (observations,1)) # x2
generated_inputs = np.column_stack((x1,x2))
noise = np.random.uniform(-1, 1, (observations,1)) # Noise
```

```python
generated_targets = 2*x1 - 3*x2 + 5 + noise
```

```python
np.savez('Files/TF_intro', inputs=generated_inputs, targets=generated_targets)
```

```python
training_data = np.load('Files/TF_intro.npz')
```

```python
input_size = 2
output_size = 1

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=output_size,
                                kernel_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1),
                                bias_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1))
                                )
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.02),loss="mean_squared_error")
```

```python
model.fit(training_data["inputs"],training_data["targets"],epochs=100,verbose=2)
```

```python
weights = model.layers[0].get_weights()[0]
weights
```

```python
bias = model.layers[0].get_weights()[1]
bias
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```
