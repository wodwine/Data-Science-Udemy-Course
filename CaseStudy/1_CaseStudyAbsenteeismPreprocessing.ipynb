```python
reason_cols = pd.get_dummies(df["Reason for Absence"],drop_first=True) # Create dummies for each reasons
reason_cols
```

```python
reason_cols["Check"] = reason_cols.sum(axis=1)
reason_cols
```

```python
reason_cols["Check"].sum(axis=0)
```

```python
reason_cols["Check"].unique()
```

```python
reason_cols.drop(["Check"],axis=1,inplace=True)
```

```python
df.drop("Reason for Absence",axis=1,inplace=True)
```

```python
reason_cols.loc[:,1:14].max(axis=1)
```

```python
reason_type_1 = reason_cols.loc[:,1:14].max(axis=1) # Group the reason into 4 gropus
reason_type_2 = reason_cols.loc[:,15:17].max(axis=1)
reason_type_3 = reason_cols.loc[:,18:21].max(axis=1)
reason_type_4 = reason_cols.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

```python
df.columns.values
```

```python
new_cols = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = new_cols # Change the name of the columns
df
```

```python
columns = ['Reason_1','Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[columns]
df
```

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod["Date"] = pd.to_datetime(df_reason_mod["Date"],format="%d/%m/%Y") # Convert the Date column to datetime
```

```python
df_reason_mod["Month"] = df_reason_mod["Date"].apply(lambda x:x.month)
df_reason_mod
```

```python
df_reason_mod["Day_Of_Week"] = df_reason_mod["Date"].apply(lambda x: x.day_of_week)
df_reason_mod
```

```python
df_reason_mod.drop("Date",axis=1,inplace=True)
```

```python
columns = ['Reason_1','Reason_2', 'Reason_3', 'Reason_4',"Month","Day_Of_Week", 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[columns]
df_reason_mod
```

```python
df_reason_date_mod = df_reason_mod.copy()
```

```python
df_reason_date_mod["Education"].value_counts()
```

```python
df_reason_date_mod["Education"] = df_reason_date_mod["Education"].map({1:0,2:1,3:1,4:1})
df_reason_date_mod
```

```python
df_preprocessed = df_reason_date_mod.copy()
df_preprocessed.head(4)
```

```python
df_preprocessed.to_csv('data/Absenteeism_preprocessed.csv', index=False)
```
