```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("data/iris_dataset.csv")
data
```

```python
x = data.copy()
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
data["Clusters"] = kmeans.predict(x)
data
```

```python
plt.scatter(data["sepal_length"],data["sepal_width"],c=data["Clusters"],cmap="rainbow");
```

```python
# Standardize
from sklearn import preprocessing

x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
# 2 Clusters
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
data["Clusters 2"] = kmeans_scaled.predict(x_scaled)
data
```

```python
plt.scatter(data["sepal_length"],data["sepal_width"],c=data["Clusters 2"],cmap="rainbow");
```

```python
# 3 Clusters
kmeans_scaled = KMeans(3)
kmeans_scaled.fit(x_scaled)
data["Clusters 3"] = kmeans_scaled.predict(x_scaled)
plt.scatter(data["sepal_length"],data["sepal_width"],c=data["Clusters 3"],cmap="rainbow");
```

```python
# 5 Clusters
kmeans_scaled = KMeans(5)
kmeans_scaled.fit(x_scaled)
data["Clusters 5"] = kmeans_scaled.predict(x_scaled)
plt.scatter(data["sepal_length"],data["sepal_width"],c=data["Clusters 5"],cmap="rainbow");
```
